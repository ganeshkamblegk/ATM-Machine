# ATM : this is an ATM simulation Project
1. crete the pin
2. changing the pin
3. check ballance
4. withdraw
5. deposit
6. exit

```python
class ATM_Machine:
    def __init__(self):
        self.pin = None
        self.balance = 0
        self.is_pin_set = False

    def create_pin(self):
        if self.is_pin_set:
            print("PIN is already set. Use change PIN option to modify.")
            return
        
        while True:
            pin = input("Enter a 4-digit PIN: ")
            if len(pin) == 4 and pin.isdigit():
                confirm_pin = input("Confirm your PIN: ")
                if pin == confirm_pin:
                    self.pin = pin
                    self.is_pin_set = True
                    print("PIN created successfully!")
                    break
                else:
                    print("PINs do not match. Please try again.")
            else:
                print("Invalid PIN. Please enter exactly 4 digits.")

    def verify_pin(self):
        if not self.is_pin_set:
            print("Please create a PIN first.")
            return False
        
        
        attempts = 3
        while attempts > 0:
            entered_pin = input("Enter your PIN: ")
            if entered_pin == self.pin:
                return True
            attempts -= 1
            print(f"Incorrect PIN. {attempts} attempts remaining.")
        
        print("Too many incorrect attempts. Please try again later.")
        return False

    def change_pin(self):
        if self.verify_pin():
            while True:
                new_pin = input("Enter new 4-digit PIN: ")
                if len(new_pin) == 4 and new_pin.isdigit():
                    confirm_pin = input("Confirm new PIN: ")
                    if new_pin == confirm_pin:
                        self.pin = new_pin
                        print("PIN changed successfully!")
                        break
                    else:
                        print("PINs do not match. Please try again.")
                else:
                    print("Invalid PIN. Please enter exactly 4 digits.")

    def check_balance(self):
        if self.verify_pin():
            print(f"Your current balance is: ₹ {self.balance:.2f}")

    def withdraw(self):
        if self.verify_pin():
            while True:
                try:
                    amount = float(input("Enter amount to withdraw: $"))
                    if amount <= 0:
                        print("Please enter a positive amount.")
                    elif amount > self.balance:
                        print("Insufficient funds.")
                    else:
                        self.balance -= amount
                        print(f"Withdrawn: ₹ {amount:.2f}")
                        print(f"New balance: ₹ {self.balance:.2f}")
                        break
                except ValueError:
                    print("Invalid amount. Please enter a number.")

    def deposit(self):
        if self.verify_pin():
            while True:
                try:
                    amount = float(input("Enter amount to deposit: $"))
                    if amount <= 0:
                        print("Please enter a positive amount.")
                    else:
                        self.balance += amount
                        print(f"Deposited: ₹ {amount:.2f}")
                        print(f"New balance: ₹ {self.balance:.2f}")
                        break
                except ValueError:
                    print("Invalid amount. Please enter a number.")

    def menu(self):
        while True:
            user_input = input("""
Please enter the number:
1. Create PIN
2. Change PIN
3. Check Balance
4. Withdraw
5. Deposit
6. Exit
Your choice: """)

            if user_input == '1':
                self.create_pin()
            elif user_input == '2':
                self.change_pin()
            elif user_input == '3':
                self.check_balance()
            elif user_input == '4':
                self.withdraw()
            elif user_input == '5':
                self.deposit()
            elif user_input == '6':
                print("Thank you for using our ATM. Goodbye!")
                break
            else:
                print("Invalid choice. Please try again.")

# Example usage
if __name__ == "__main__":
    atm = ATM_Machine()
    atm.menu()
```
